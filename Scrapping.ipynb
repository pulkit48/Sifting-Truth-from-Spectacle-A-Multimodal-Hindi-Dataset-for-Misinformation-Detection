```python
def clean(t2):
  t2=t2[:]
  temp=""
  p=0
  for i in t2:
    if i.strong:
      i.strong.decompose()
    if i.span:
      i.span.decompose()

    temp+=i.get_text().replace('\n',' ')

  return temp
```

```python
def eval(src_value1,ttl):

    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    src_value2=requests.get(src_value1,headers=headers).text

    soup2 = BeautifulSoup(src_value2, 'html.parser')

    t2=clean(soup2.find_all('p'))

    img=soup2.find('div',id="oi-vox")
    # ttl=soup2.find('h1').text
    if(img!=None):
      img=img.find('picture').find('img')['src']
      Image_link.append(img)
      News_description.append(t2.strip())
      article_link.append(src_value1)
      Rating.append("True")
      title.append(ttl)
```

```python
# Function to scroll to the bottom of the page
def scroll_to_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)  # You may adjust the sleep duration based on the website's loading time
```
